```python
#7.4 함수로 묶기
#우리가 필요한 제품명, 가격, 링크 세 가지 정보는 모두 <div class = "item_info_cont">에 저장되어 있음
# <div class = "item_info_cont"> 를 뽑은 후 3가지 정보를 한 번에 뽑는 로직을 함수로 묶어
# {"name":"~~~","price":"~~~", "link":"~~~"} 와 같이 딕셔너리 형태로 출력
'''
ul = bs_obj.find("ul", {"class":"item_box line_4"})
items = ul.findAll("div", {"class":"item_info_cont"})

for item in items:
  strong_tag = item.find("strong", {"class":"item_name"})
  print(strong_tag.text)
'''

ul = bs_obj.find("ul", {"class":"item_box line_4"})
items = ul.findAll("div", {"class":"item_info_cont"})

def get_product_info(item):
  strong_tag = item.find("strong", {"class":"item_name"})
  return strong_tag.text

for item in items:
  product_info = get_product_info(item)
  print(product_info)
```

```python
# 7.5 제품명, 가격 함께 뽑기, 7.6 딕셔너리로 여러 개 값 리턴하기
ul = bs_obj.find("ul", {"class":"item_box line_4"})
items = ul.findAll("div", {"class":"item_info_cont"})

def get_product_info(item):
  name = item.find("strong", {"class":"item_name"})
  price = item.find("span", {"class":"item_price"})
  return {"name":name.text, "price":price.text}

for item in items:
  product_info = get_product_info(item)
  print(product_info)
```

```python
# 7.7 링크 뽑아내기
ul = bs_obj.find("ul", {"class":"item_box line_4"})
items = ul.findAll("div", {"class":"item_info_cont"})

def get_product_info(item):
  name = item.find("strong", {"class":"item_name"})
  price = item.find("span", {"class":"item_price"})
  link = item.find("a")['href'] # a태그에서 href 속성에 들어있는 링크 추출
  return {"name":name.text, "price":price.text, "link":link}

for item in items:
  product_info = get_product_info(item)
  print(product_info)
```

```python
# 7.8 한 페이지 데이터를 한 덩어리로 만들기
ul = bs_obj.find("ul", {"class":"item_box line_4"})
items = ul.findAll("div", {"class":"item_info_cont"})

def get_product_info(item):
  name = item.find("strong", {"class":"item_name"})
  price = item.find("span", {"class":"item_price"})
  link = item.find("a")['href'] # a태그에서 href 속성에 들어있는 링크 추출
  return {"name":name.text, "price":price.text, "link":link}

#리스트 축약
product_info_list = [get_product_info(item) for item in items]

print(product_info_list)
```

```python
# 7.9 한 페이지를 호출하는 부분을 함수로 감싸기
# 지금까지 작성한 로직은 특정 URL을 입력하면 그 url에 있는 상품 정보를 리스트에 넣어주는 기능
import requests
import bs4

def get_product_info(item):
  name = item.find("strong", {"class":"item_name"})
  price = item.find("span", {"class":"item_price"})
  link = item.find("a")['href'] # a태그에서 href 속성에 들어있는 링크 추출
  return {"name":name.text, "price":price.text, "link":link}


def get_page_products(url):
  result = requests.get(url)
  bs_obj = bs4.BeautifulSoup(result.content, "html.parser")
  ul = bs_obj.find("ul",{"class": "item_box line_4"})
  items = ul.findAll("div", {"class":"item_info_cont"})
  product_info_list = [get_product_info(item) for item in items]
  return product_info_list

#url만 바꾸어주면 그 페이지에 있는 상품 정보를 모두 받아올 수 있음
url = "https://www.sidmool.com/shop/shopbrand.html?type=M&xcode=022&mcode=002&viewtype=gallery"
page_products = get_page_products(url)
print(page_products)
```

```python
#7.10 1페이지부터 5페이지까지 데이터 받아오기(여기서는 2페이지까지)
#각 페이지의 주소(url)를 알아야 함

import requests
import bs4

def get_product_info(item):
  name = item.find("strong", {"class":"item_name"})
  price = item.find("span", {"class":"item_price"})
  link = item.find("a")['href'] # a태그에서 href 속성에 들어있는 링크 추출
  return {"name":name.text, "price":price.text, "link":link}


def get_page_products(url):
  result = requests.get(url)
  bs_obj = bs4.BeautifulSoup(result.content, "html.parser")
  ul = bs_obj.find("ul",{"class": "item_box line_4"})
  items = ul.findAll("div", {"class":"item_info_cont"})
  product_info_list = [get_product_info(item) for item in items]
  return product_info_list

urls =[
    "https://www.sidmool.com/shop/shopbrand.html?type=X&xcode=022&mcode=002&sort=&page=1", #1페이지
    "https://www.sidmool.com/shop/shopbrand.html?type=X&xcode=022&mcode=002&sort=&page=2" #2페이지
]

for page_number in range(0,2):
  page_products = get_page_products(urls[page_number])
  print(len(page_products), page_products)
```
